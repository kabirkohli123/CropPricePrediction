```python
X = data[['state_name', 'district_name', 'market_name', 'commodity_name',
    'month_column', 'day']]
Y = data['modal_price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
X_train_scaled = np.reshape(X_train_scaled, (X_train_scaled.shape[0], X_train_scaled.shape[1], 1))
X_test_scaled = np.reshape(X_test_scaled, (X_test_scaled.shape[0], X_test_scaled.shape[1], 1))
```

```python
from sklearn import metrics
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
```

```python
reg = RandomForestRegressor(max_depth = 1000, random_state =0)
reg.fit(X_train,Y_train)
```

```python
Y_pred = reg.predict(X_test)
```

```python
print("The Predicted value of Y is :")
print(Y_pred[:10])
print("The test value of Y is :")
print(np.array(Y_test[:10]))
print("The accuracy of Random Forest is :")
print(reg.score(X_test,Y_test))
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
```

```python
model = keras.Sequential([keras.layers.Flatten(input_shape=(6,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
                         ])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
history = model.fit(X_train_scaled, Y_train, epochs=10, batch_size=32, validation_data=(X_test_scaled, Y_test))
```

```python
Y_pred_lstm = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import r2_score

# y_true: actual values, y_pred: predicted values
r2 = r2_score(Y_test, Y_pred_lstm)
print(f"R² score: {r2}")
```

```python
from sklearn.metrics import mean_squared_error
mse_lstm = mean_squared_error(Y_test, Y_pred_lstm)
print(f"LSTM Model Mean Squared Error: {mse_lstm}")
```
